# Full Observed BCM with Leaders

### The model
In this notebook I estimate $\epsilon = (\epsilon^+_F, \epsilon^+_L, \epsilon^-_F, \epsilon^-_L)$ in a BCM with backfire effect with leaders and followers.

- The population is divided between Leaders and Followers.
- Leaders are ruled by the parameters $\epsilon^+_L$, $\epsilon^-_L$, $\mu^+_L$, $\mu^-_L$.
- $R \in \{0,1\}^{N}$ is the vector of roles. If $u$ is a leader, $R_u = 1$, otherwise $R_u = 0$.
- Followers are ruled by the parameters $\epsilon^+_F$, $\epsilon^-_F$, $\mu^+_F$, $\mu^-_F$.
- $\mu = (\mu^+_F, \mu^+_L, \mu^-_F, \mu^-_L)$ are observed.

Edges have the form $e_j = (u,v,s^+,s^-,t)$, meaning that if two users interact, they can have both a positive interaction and a negative interaction.

$
P(s^+_j = 1 \mid \epsilon, e_j, X, R) = \sigma(\rho \cdot (\epsilon^+_{R_v} - \lvert X^t_u - X^t_v \rvert ) )
$

$
P(s^-_j = 1 \mid \epsilon, e_j, X, R) = \sigma(-\rho \cdot (\epsilon^-_{R_v} - \lvert X^t_u - X^t_v \rvert ) )
$

After the interaction, $v$ update its opinion accordingly

$X^{t+1}_v += \mu^+_{R_v} \cdot s^+_j (X^t_u - X^t_v) - \mu^-_{R_v} \cdot s^-_j (X^t_u - X^t_v) $

Finally, $X^{t+1}$ is clipped in $[0,1]$

### SVI 

To this end, we estimate $\epsilon$ with Stochastic Variational Inference with Normalizing Flows.

- We use Normalizing Flows (Sylvester Flows) as variational distribution.
- At each step, a distribution $q_\phi$ is extracted from the variational family.
- $\theta$ is sampled  from $q_\phi$.
- $\epsilon^+ = \sigma(\theta) / 2$, $\epsilon^- = \sigma(\theta) / 2 + 0.5$
- The optimizer minimizes the ELBO, that is equivalent to find $\phi$ that minimize $KL(p \| q_\phi)$

### Normalizing Flows

```python
## assume to observe both the roles, the opinions, and the interactions

class NormalizingFlow(nn.Module):
    def __init__(self,dim,
                      n_flows,
                     base_dist=lambda dim:distributions.Uniform(torch.zeros(dim), torch.ones(dim)),
                     flow_type=lambda kwargs:distributions.transforms.RadialFlow(**kwargs),
                     args={'flow_args':{'dim':4}}):
        super(NormalizingFlow, self).__init__()
        self.dim = dim
        self.n_flows = n_flows
        self.base_dist = base_dist(dim)
        self.uuid = np.random.randint(low=0,high=10000,size=1)[0]
        
        
        """
        If the flow needs an autoregressive net, build it for every flow
        """
        if 'arn_hidden' in args:
            self.arns = nn.ModuleList([AutoRegressiveNN(dim,
                                                        args['arn_hidden'],
                                                        param_dims=[self.dim]*args['n_params']) for _ in range(n_flows)])
    
        """
        Initialize all flows
        """
        self.nfs = []
        for f in range(n_flows):
            if 'autoregressive_nn' in args['flow_args']:
                args['flow_args']['autoregressive_nn'] = self.arns[f]
            nf = flow_type(args['flow_args'])
            self.nfs.append(nf)

        """
        This step assumes that nfs={f_i}_{i=1}^N and that base_dist=N(0,I)
        Then, register the (biejctive) transformation Z=nfs(eps), eps~base_dist
        """
        self.nf_dist = distributions.TransformedDistribution(self.base_dist, self.nfs)
        
        self._register()
        
    def _register(self):
        """
        Register all N flows with Pyro
        """
        for f in range(self.n_flows):
            nf_module = pyro.module("%d_nf_%d" %(self.uuid,f), self.nfs[f])

    
    def target(self,X,roles,u,v,s_plus_, s_minus_,t,rho):
        
        dist = self.nf_dist
        theta = pyro.sample("latent", dist)
        epsilons = torch.sigmoid(theta) / 2
        epsilon_plus_F, epsilon_plus_L = epsilons[:2]
        epsilon_minus_F, epsilon_minus_L = epsilons[2:] + .5
        
        diff_X = X[t,u] - X[t,v]
        roles_v = roles[v]
        
        epsilon_plus_r = (1 - roles_v) * epsilon_plus_F + roles_v * epsilon_plus_L
        epsilon_minus_r = (1 - roles_v) * epsilon_minus_F + roles_v * epsilon_minus_L
        
        kappas_plus = BC_roles.kappa_plus_from_epsilon(epsilon_plus_r, diff_X, rho)
        kappas_minus = BC_roles.kappa_minus_from_epsilon(epsilon_minus_r, diff_X, rho)
        kappas_ = torch.cat([kappas_minus, kappas_plus])
        s_ = torch.cat([s_minus_, s_plus_])
        
        #with pyro.plate("data", s_plus_.shape[0]):
            #pyro.sample("obs", distributions.Bernoulli(probs = kappas_minus), 
            #            obs = s_minus_)
            #pyro.sample("obs_plus", distributions.Bernoulli(probs = kappas_plus), obs = s_plus_)
        with pyro.plate("data", s_.shape[0]):
            pyro.sample("obs", distributions.Bernoulli(probs = kappas_), obs = s_)


    def model(self,X,roles,u,v,s_plus_, s_minus_,t,rho):
        """
        q(z|x), once again x is not required
        
        1. Sample Z ~ nfs(eps), eps ~ N(0,I)
        
        This is the NN being trained
        """
        self._register()
        # print("nf_event", self.nf_dist.event_shape)
        # print("nf_batch", self.nf_dist.batch_shape)
        pyro.sample("latent", self.nf_dist)

    def sample(self,n):
        """
        Sample a batch of (n,dim)
        
        Bug: in IAF and IAFStable, the dimensions throw an error (todo)
        """
        return self.nf_dist.sample(torch.Size([n]))
    
    def log_prob(self,z):
        """
        Returns log q(z|x) for z (assuming no x is required)
        """
        return self.nf_dist.log_prob(z)
```

### Train

```python
dim = 4

flow = distributions.transforms.Sylvester #2000 epochs. loss 135
base_dist_n = lambda dim:distributions.Normal(torch.zeros(dim), torch.ones(dim)) 
base_dist_u = lambda dim:distributions.Uniform(torch.zeros(dim), torch.ones(dim)) 

n_flows = 32

if flow.__name__ in ['Sylvester']:
    args = {'flow_args':{'input_dim':dim,
                         'count_transforms':32}}
```

```python
base_dist_s = lambda dim:distributions.Uniform(torch.zeros(dim), torch.ones(dim))
u,v,s_plus,s_minus,t = BC_roles.convert_edges_uvst(edges)

s_plus_ = s_plus.to(torch.float32)
s_minus_ = s_minus.to(torch.float32)

nf_obj = NormalizingFlow(dim=dim,
                         n_flows=n_flows,
                         base_dist=base_dist_s,
                         flow_type=lambda kwargs:flow(**kwargs),
                         args=args)
```

```python
n_steps = 200
```

```python
adam_params = {"lr": 0.05, "betas": (0.90, 0.999)}
optimizer = Adam(adam_params)

t0 = time()

# setup the inference algorithm
svi = SVI(nf_obj.target, nf_obj.model, optimizer, loss=Trace_ELBO())
# dist = p_z() # true distribution
# do gradient steps
losses = []
for step in tqdm(range(n_steps)):
    loss = svi.step(X,sim.roles,u,v,s_plus_, s_minus_,t,rho) # analogous to opt.step() in PyTorch
    losses.append(loss)

t1 = time()
tot_time = t1 - t0

fig,axes = sbp(figsize = (2.5,2.5))
axes.plot(losses)
```

### Complete training and analysis

```python
def train(X, edges, roles, dist_name = "Normal", n_flows = 8, 
          count_transform = 12, n_steps = 100, lr = 0.05,
          print_loss = True, hide_progress = False, 
          print_time = True, rho = 16
         ):
    dim = 4
    flow = distributions.transforms.Sylvester #2000 epochs. loss 135
    if dist_name == "Normal":
        base_dist_s = lambda dim:distributions.Normal(torch.zeros(dim), torch.ones(dim)) 
    elif dist_name == "Uniform":
        base_dist_s = lambda dim:distributions.Uniform(torch.zeros(dim), torch.ones(dim)) 
    args = {'flow_args':{'input_dim':dim,
                         'count_transforms':count_transform}}
    
    nf_obj = NormalizingFlow(dim=dim,
                             n_flows=n_flows,
                             base_dist=base_dist_s,
                             flow_type=lambda kwargs:flow(**kwargs),
                             args=args)
    
    
    u,v,s_plus,s_minus,t = BC_roles.convert_edges_uvst(edges)
    s_plus_ = s_plus.to(torch.float32)
    s_minus_ = s_minus.to(torch.float32)
    
    adam_params = {"lr": lr, "betas": (0.90, 0.999)}
    optimizer = Adam(adam_params)

    t0 = time()
    
    # setup the inference algorithm
    svi = SVI(nf_obj.target, nf_obj.model, optimizer, loss = Trace_ELBO())
    
    
    losses = []
    for step in tqdm(range(n_steps), disable = hide_progress):
        loss = svi.step(X, roles, u, v, s_plus_, s_minus_, t, rho) # analogous to opt.step() in PyTorch
        losses.append(loss)
    
    t1 = time()
    tot_time = t1 - t0
    
    if print_loss:
        fig,axes = sbp(figsize = (2.5,2.5))
        axes.plot(losses)
        fig.show()
        
    if print_time:
        print(tot_time, "s")
        
    return nf_obj
```

```python
def print_median_nf(samples, targets, eps_names = ["$\epsilon^+_F$","$\epsilon^+_L$","$\epsilon^-_F$","$\epsilon^-_L$"]):
    for j in range(4):
        print(eps_names[j], round(targets[j].item(), 3), round(samples[j].median().item(), 3)) 
    

def print_uni_nf(samples, targets, figsize = (12,4)):
    fig, axes = sbp(ncols = 4, figsize = figsize)
    
    for j in range(4):
        p = sns.kdeplot(samples[j], ax = axes[j])
        ymax_p = p.lines[0].get_data()[1].max()
        axes[j].vlines(x = targets[j], ymin = 0, ymax = ymax_p, colors = "red")
    fig.show()
    
def print_pairs_nf(samples, targets, figsize = (13,13)):
    fig, axes = sbp(ncols = 4, nrows = 4, figsize = figsize)
    
    for i, var1 in enumerate(samples):
        for j, var2 in enumerate(samples):
            axes[i,j].set_title(eps_names[i] +  " " + eps_names[j])
            axes[i,j].set_ylabel("")
            if i == j:
                p = sns.kdeplot(var1, ax = axes[i,j])
                ymax_p = p.lines[0].get_data()[1].max()
                axes[i,j].vlines(x = targets[i], ymin = 0, ymax = ymax_p, colors = "red")
                axes[i,j].set_ylabel("")
            else:
                sns.kdeplot(x = var1, y = var2, ax = axes[i,j], fill = True)
                axes[i,j].scatter(x = targets[i], y = targets[j], color = "red")
        
    fig.show()
```

```python
def nf_analysis(nf_obj, targets, print_median = True, print_pairs = True, print_uni = False, 
                eps_names = ["$\epsilon^+_F$","$\epsilon^+_L$","$\epsilon^-_F$","$\epsilon^-_L$"]):
    
    epsilon_plus_F_s, epsilon_plus_L_s, epsilon_minus_F_s, epsilon_minus_L_s = (torch.sigmoid(nf_obj.sample(1000)) / 2).T
    epsilon_minus_F_s += .5
    epsilon_minus_L_s += .5
    
    samples = [epsilon_plus_F_s, epsilon_plus_L_s, epsilon_minus_F_s, epsilon_minus_L_s]

    
    if print_median:
        print_median_nf(samples, targets)
    
    if print_uni:
        print_uni_nf(samples, targets)
    
    if print_pairs:
        print_pairs_nf(samples, targets)
```

```python
N = 100
T = 2 ** 12
print("T", T)
edge_per_t = 9
# epsilon = (epsilon_F, epsilon_L)
epsilon_plus = torch.Tensor([0.45, 0.05])
epsilon_minus = torch.Tensor([0.55, 0.95])
mu_plus = torch.Tensor([0.05, 0.1])
mu_minus = torch.Tensor([0.05, 0.1])
rho = 32
initial_leaders = 15

targets = (epsilon_plus[0], epsilon_plus[1], epsilon_minus[0], epsilon_minus[1])

sim.initialize_simulator(N, T, edge_per_t, initial_leaders)
X, edges = sim.simulate_trajectory((epsilon_plus, epsilon_minus, mu_plus, mu_minus, rho))
```

```python
nf_est = train(X, edges, sim.roles)
```

```python
nf_analysis(nf_est, targets)
```
